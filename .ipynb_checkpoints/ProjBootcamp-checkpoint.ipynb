```python
sns.set_style("whitegrid")


plt.figure(figsize=(12, 8))

df[df.type=='Movie'].release_year.plot(bins=35, kind='hist', color='orange', 
                                       label='Movie Release', alpha=0.2)
df[df.type=='TV Show'].release_year.plot(kind='hist', color='blue', 
                                       label='Series Release', alpha=0.2)
plt.legend()
plt.xlabel("Release Year")
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
colors = sns.color_palette("muted")

# atur datanya
df_pie = df[['type']]
df_pie = df_pie.groupby(['type'])['type'].count().reset_index(name="type_count")

# konversi data ke chart
plt.pie(df_pie['type_count'], labels=df_pie['type'], colors=colors, autopct = '%0.2f%%')
plt.show()
```

```python
show_origin= df.groupby(['country','type'])[ 'type'].count().reset_index(name='show_count')
show_origin

```

```python
top20_country=show_origin.query("`show_count` >= 81")
top20_country
```

```python
sns.barplot(x =top20_country.reset_index()['country'], y=top20_country.reset_index()['show_count'],
            hue =  top20_country.reset_index()['type']);
plt.title(' overall_top20_country', fontsize=16)
plt.xlabel('country', fontsize=16);
plt.ylabel('show_count', fontsize=16);
sns.set(rc = {'figure.figsize':(20,10)})
plt.xticks(rotation = 45);
```

```python

df['duration2'] = df['duration'].str.findall(r'\d+').str.join('').astype(int)
df


```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
counts = df.groupby(['duration2','type','rating'])['duration2'].count().reset_index(name='counts')
counts
counts1 = df.groupby(['duration2','type', 'rating'])['duration2'].count().reset_index(name='counts1')
counts1
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
ax = sns.barplot(x='duration2', y='counts1', data=counts1.loc[counts1['type'] == 'TV Show'])
plt.ylabel("TV Shows Count")
plt.xlabel("TV Shows Seasons")



#sns.scatterplot(data=counts1.loc[counts1['type'] == 'TV Show'], x="duration2", y="counts1")
#sns.rugplot(data=counts1.loc[counts1['type'] == 'TV Show'], x="duration2", y="counts1")


```

```python
#ay = sns.barplot(x='counts1', y='duration2', data=counts1.loc[counts1['type'] == 'Movie'])

#sns.ecdfplot(data=counts1, x="release_year", hue="type" )
#sns.lmplot(data=counts1.loc[counts1['type'] == 'TV Show'], x="release_year", y="counts1", hue="rating")
sns.relplot(
    x="rating", 
    y="counts", 
    hue="rating",
    alpha=.8, 
    palette="muted",
    height=6, 
    data=counts.loc[counts['type'] == 'Movie']
)

plt.title("Example Relational Scatter Plot");
plt.ylabel("Movies Count")
plt.xlabel("Movies Duration")

```
