```python
show_origin= df.groupby(['country'])['type'].count().reset_index(name='show_count')
show_origin
```

```python
top10_country=show_origin.query("`show_count` >= 150")
top10_country.reset_index()
```

```python
plt.rcdefaults()
# Create figure
fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(aspect="equal"))
colors = sns.color_palette("husl", 8)
plt.pie(top10_country['show_count'], labels=top10_country['country'], colors=colors, autopct = '%0.1f%%')

# Add a white circle to center of pie chart
centre_circle = plt.Circle((0,0),0.80,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Top 10 countries',fontsize=14)
plt.show()
```

```python
# seaborn catplot (horizontal grouped bar chart)
plt.figure(figsize=(8,6))
ax = sns.histplot(y="rating", data=df,
                  hue="type",
                  edgecolor=".6",
                  legend=True, palette=sns.color_palette('hls',2))

plt.xlabel('Number Of Movie and TV Shows in each Rate',fontsize = 12)
plt.ylabel('Rating',fontsize = 12)

ax.set_title('Rating based on Type',fontsize=16)
plt.show()
```
