```python

import seaborn as sns
```

```python
sns.set_style("whitegrid")


plt.figure(figsize=(12, 8))

df[df.type=='Movie'].release_year.plot(bins=35, kind='hist', color='orange', 
                                       label='Movie Release', alpha=0.2)
df[df.type=='TV Show'].release_year.plot(kind='hist', color='blue', 
                                       label='Series Release', alpha=0.2)
plt.legend()
plt.xlabel("Release Year")
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
colors = sns.color_palette("muted")

# atur datanya
df_pie = df[['type']]
df_pie = df_pie.groupby(['type'])['type'].count().reset_index(name="type_count")

# konversi data ke chart
plt.pie(df_pie['type_count'], labels=df_pie['type'], colors=colors, autopct = '%0.2f%%')
plt.show()
```

```python
show_origin= df.groupby(['country','type'])[ 'type'].count().reset_index(name='show_count')
show_origin

```

```python
top20_country=show_origin.query("`show_count` >= 81")
top20_country
```

```python
sns.barplot(x =top20_country.reset_index()['country'], y=top20_country.reset_index()['show_count'],
            hue =  top20_country.reset_index()['type']);
plt.title(' overall_top20_country', fontsize=16)
plt.xlabel('country', fontsize=16);
plt.ylabel('show_count', fontsize=16);
sns.set(rc = {'figure.figsize':(20,10)})
plt.xticks(rotation = 45);
```
